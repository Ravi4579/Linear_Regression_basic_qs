```python
import statsmodels.formula.api as smf
model=smf.ols("Salary~YearsExperience",data=df).fit()
```

```python
model.summary()
```

Predicting the salary acc to years

```python
print(model.tvalues, model.pvalues, model.rsquared, model.rsquared_adj)
```

```python
predictions = model.predict(df["YearsExperience"])
```
